```python
import matplotlib.pyplot as plt
%matplotlib inline

exp_values = [1400,600,300,410,250]
exp_labels = [ "Home Rent" , "Food" , "Phone Bill", "Car" , "Other Utilities" ]

plt.pie ( exp_values , labels=exp_labels , radius=1.2 , autopct="%0.2f%%" , shadow=True , 
         explode=[0,0,0,0,0] , startangle=-180 , labeldistance=1.1 )

plt.axis("equal")
plt.show()
```
